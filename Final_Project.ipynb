**Tugas Final Project - Data Analytics**

Teman-teman dipersilahkan untuk membuat aplikasi Streamlit untuk Data Visualization atau AI/ML dan juga menggunakan data yang lain selain data yang digunakan dikelas.

Tugas streamlit codes perlu di deploy ke streamlit cloud dan bisa share link aplikasi streamlitnya ke WA group atau info@dtsense.id,
jika tidak bisa mendeploy, streamlit codes bisa dishare/dikirimkan ke info@dtsense.id atau di share di WA Group dan bisa disertakan dengan nama peserta.

Durasi tugas up to 2 weeks setelah workshop selesai.
Sertifikat akan dibagikan H+7 setelah pengumpulan.

Jika, memerlukan waktu lebih untuk menyelesaikan final project, bisa dikabarkan ke instruktur/admin.
